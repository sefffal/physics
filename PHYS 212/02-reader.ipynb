```python
def x(w, t, delta):
    return np.cos(w*t + delta)
```

```python
def y(w, t, delta):
    return np.cos(w*t + delta)
```

## Example 1
### Superimposed vibrations of different frequencies: Beats Figure 2-4

The first thing we do is make our own version of Figure 2-4. We generate two waves x1 and x2 with frequencies 450 and 150 Hz. In this first panel we plot the two waves and in the bottom panel we add them. They are commensurate and therefore the resultant wave is periodic. 

```python
time = np.linspace(start=0,stop=0.02,num=1000)
x1 = x(450*2*np.pi,time, 0.0)
x2 = x(150*2*np.pi,time, 0.0)
```

```python
plt.figure(num=0, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(211)
plt.plot(x1,'b',x2,'r') 
plt.xlim([0, 1000])
plt.ylabel('Amplitude / Arb. Units',fontsize=14)
plt.xlabel('Time / S',fontsize=14)
plt.subplot(212)
plt.plot(x1+x2,'g') 
plt.xlim([0, 1000])
plt.ylabel('Amplitude / Arb. Units',fontsize=14)
plt.xlabel('Time / S',fontsize=14)
plt.show()
```

## Example 2
### Superimposed vibrations of different frequencies
### Beats Figure 2-6

When two waves are added with frequencies that are close beats are generated. It is shown in the text that the sum of two cos functions can be written as a product of cos functions. One of the cosines can be considered to be the carrier wave - it has a frequency equal to the mean of the two components. The other cosine can be thought to modulate the carrier wave at a frequency equal to half of the difference of the two component frequencies. This produces a beat frequency that, as the frequency of the two component waves gets closer, gets slower. Beats can be used to tune stringed instruments. An example can be found in the following YouTube video: https://www.youtube.com/watch?v=srNqPEJJCLU

```python
time = np.linspace(start=0,stop=0.02,num=1000)
x1 = x(700*2*np.pi,time, 0.0)
x2 = x(600*2*np.pi,time, 0.0)
```

```python
plt.figure(num=0, figsize=(10, 8), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(211)
plt.plot(x1,'b',x2,'r') 
plt.xlim([0, 1000])
plt.ylabel('Amplitude / Arb. Units',fontsize=14)
plt.xlabel('Time / S',fontsize=14)
plt.subplot(212)
plt.plot(x1+x2,'g') 
plt.xlim([0, 1000])
plt.ylabel('Amplitude / Arb. Units',fontsize=14)
plt.xlabel('Time / S',fontsize=14)
plt.show()
```

## Example 3
### Perpendicular motions with equal frequencies Figure 2-10

Below you will some example code that will allow you to recreate Figure 2-10 from the textbook.  

Now we divide up $2\pi$ radians into 1000 parts and use this as our time variable. 

```python
t = np.linspace(0,2*np.pi,1000)
```

What I would like you to do in this example is add the correct phase difference to each subplot. The phase difference (delta) for the first subplot has been set equal to zero. The phase difference for the second subplot is pi/4. Both of these are correct but all the others are wrong. If you could change the value of delta for each subplot to the correct value, you will be able to reproduce the Figure 2-10 in the textbook. 

```python
def makeasubplot(index,delta):
    plt.subplot(index)
    plt.plot(x(1,t,0.0), y(1,t,delta))
    plt.axis('equal')
```

```python
plt.figure(num=1, figsize=(12, 12), dpi=80, facecolor='w', edgecolor='k')

makeasubplot(331,0)
plt.title('$\delta = 0$',fontsize=18)
makeasubplot(332,np.pi/4)
plt.title('$\pi/4$',fontsize=18)
makeasubplot(333,np.pi/4) #CHANGE 
plt.title('$\pi/2$',fontsize=18)
makeasubplot(334,np.pi/4) #CHANGE 
plt.title('$3\pi/4$',fontsize=18)
makeasubplot(335,np.pi/4) #CHANGE 
plt.title('$\pi$',fontsize=18)
makeasubplot(336,np.pi/4) #CHANGE 
plt.title('$5\pi/4$',fontsize=18)
makeasubplot(337,np.pi/4) #CHANGE 
plt.title('$3\pi/2$',fontsize=18)
makeasubplot(338,np.pi/4) #CHANGE 
plt.title('$7\pi/4$',fontsize=18)
makeasubplot(339,np.pi/4) #CHANGE 
plt.title('$2\pi$',fontsize=18)

plt.show()
```

### Lissajous Figures 2-14

We can also change the ratio of the frequencies, to generate each of the panels in Figure 2-14, by simply changing the integers 1 and 2. Note that to get them to appear like the book I have to interchange the indexes. 1:3 becomes 3:1 and I had to flip the x-axis by -1. Then they work out. The first thing you could try is the sequence (1,1,0), (1,2,0), (1,3,0) etc. That will generate the first column in Figure 2-14. Then you could try (1,1,np.pi/4), (1,2,np.pi/4), (1,3,np.pi/4) etc.

```python
def plotpanelfromfigure(i,j,delta):
    plt.figure(num=0, figsize=(6, 6), dpi=80, facecolor='w', edgecolor='k')
    plt.plot(-x(j,t,0),y(i,t, delta)) 
    plt.axis('equal')
    plt.ylabel('y',fontsize=16)
    plt.xlabel('x',fontsize=16)
    plt.show()
```

```python
plotpanelfromfigure(1,3,0)
```

### From Glowing Python

http://glowingpython.blogspot.ca/2011/12/lissajous-curves.html

```python
from numpy import sin,pi,linspace
from pylab import plot,show,subplot

a = [1,3,5,3] # plotting the curves for
b = [1,5,7,4] # different values of a/b
delta = pi/2
t = linspace(-pi,pi,300)

plt.figure(num=0, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')

for i in range(0,4):
    x = sin(a[i]*t + delta)
    y = sin(b[i]*t)
    plt.subplot(2,2,i+1)
    plt.axis('equal')
    plt.plot(x,y)

plt.show()
```

$\begin{eqnarray}
x_0 =& A \sin(\omega_0 t + \phi_0), \;\; \phi_0 =& 0 \\
x_1 =& A \sin(\omega_0 t + \phi_1), \;\; \phi_0 =& 45^\circ \\
x_2 =& A \sin(\omega_0 t + \phi_2), \;\; \phi_0 =& 75^\circ \\
\end{eqnarray}$

```python
%pylab inline

def plotvibrations(x1, x2, t):
    plt.figure(num=0, figsize=(10,12), dpi=80, facecolor='w', edgecolor='k')
    plt.subplot(211)
    plt.plot(t, x1,'b') 
    plt.plot(t, x2,'r') 
    plt.ylabel('Amplitude / Arb. Units',fontsize=14)
    plt.xlabel('Time / S',fontsize=14)
    plt.subplot(212)
    plt.plot(t, x1+x2,'g') 
    plt.ylabel('Amplitude / Arb. Units',fontsize=14)
    plt.xlabel('Time / S',fontsize=14)
    plt.show()
def plotlissajousfigure(x, y):
    plt.figure(num=0, figsize=(6, 6), dpi=80, facecolor='w', edgecolor='k')
    plt.plot(x, y) 
    plt.axis('equal')
    plt.ylabel('y',fontsize=16)
    plt.xlabel('x',fontsize=16)
    plt.show()
```

```python
t = linspace(0, pi, 500)
x = cos(2*t)
y = cos(t)
plotlissajousfigure(x,y)
```

```python
t = linspace(0, pi/5, 500)
x = 10*cos(5*pi*t)
y = 10*cos(10*pi*t+pi/3)
plotlissajousfigure(x,y)
```

```python
w = 2*pi
t= linspace(0, pi/2, 1000)

x1 = 0.25*sin(w*t)
x2 = 0.20*sin(w*t+deg2rad(45))
x3 = 0.15*sin(w*t+deg2rad(45+30))
sup_posit = x1+x2+x3

# Pick out the amplitude and phase shift
amp = max(sup_posit)
phi = t[sup_posit==max(sup_posit)][0]

plot(t, sup_posit)
text(0.4, 0.2,
     '$A={:.2f}$\n$\phi={:.2f}$'.format(amp, phi),
     fontsize=16
);
title('Addition of three waves of equal frequency')
xlabel('Displacement (Arb. units)')
xlabel('Time (Arb. units)');
```
